## Demonstrating the word replacement task using function

```python
def replce_str(str_to_be_replaced,new_string,file_name):
    with open(file_name,"r+") as f:
        s=f.read()
        new_sentance=s.replace(str_to_be_replaced,new_string)
    with open(file_name,"w") as f:
        f.write(new_sentance)
    
replce_str("placement","screening","example.txt")  
```

## Demonstrating word replacement task using abstractmethod and multiple inheritance

```python
from abc import ABC,abstractmethod

class print_sentence(ABC):
    @abstractmethod
    def print_sentence(self):
        pass



class str_intiation():
    #constructor
    def __init__(self,str_to_be_replaced,new_string,file_name):
        self.str_to_be_replaced=str_to_be_replaced
        self.new_string=new_string
        self.file_name=file_name
        
        
class str_replacement(str_intiation,print_sentence):
    
    def replce_str(self):
        with open(self.file_name,"r+") as f:
            s=f.read()
            new_sentance=s.replace(self.str_to_be_replaced,self.new_string)
        with open(self.file_name,"w") as f:
            f.write(new_sentance)
    #Enforcing child class to print the updated sentances using abstractmethod in parent class        
    def print_sentence(self):
        with open(self.file_name,"r+") as f:
            print(f.read())
            
object_=str_replacement("placement","screening","example.txt")
print("Before replacement")
object_.print_sentence()
object_.replce_str()
print("After replacement")
object_.print_sentence()
```

## Demonstrating word replacement task using Decorator

```python
def str_replacement(func):
    def actual(str_to_be_replaced,new_string,file_name):
        print("Before replacement")
        with open(file_name,"r+") as f:
                s=f.read()
                print(s)
                new_sentance=s.replace(str_to_be_replaced,new_string)
        with open(file_name,"w") as f:
                f.write(new_sentance)
        out=func(str_to_be_replaced,new_string,file_name)
        return out
    return actual

@str_replacement
def print_out(str_to_be_replaced,new_string,file_name):
    print("After replacement")
    with open(file_name,"r+") as f:
        print(f.read())
    

print_out("screening","placement","example.txt")
```
